```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import ipywidgets as widgets
from IPython.display import display
```

```python
def ode_visualizer(a=-1, y0=1, t_end=10):

    # Differential Equation
    def ode(t, y):
        return a * y

    # Solve ODE
    t_span = (0, t_end)
    t_eval = np.linspace(0, t_end, 500)

    solution = solve_ivp(
        ode,
        t_span,
        [y0],
        t_eval=t_eval
    )

    # Plot
    plt.figure(figsize=(10,5))
    plt.plot(solution.t, solution.y[0],
             linewidth=3,
             label=f"dy/dt = {a}y")

    plt.title("ODE Solver Visualizer")
    plt.xlabel("Time (t)")
    plt.ylabel("y(t)")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
widgets.interact(
    ode_visualizer,
    a=widgets.FloatSlider(
        value=-1,
        min=-5,
        max=5,
        step=0.1,
        description='a'
    ),
    y0=widgets.FloatSlider(
        value=1,
        min=-10,
        max=10,
        step=0.1,
        description='y0'
    ),
    t_end=widgets.IntSlider(
        value=10,
        min=1,
        max=50,
        step=1,
        description='Time'
    )
);
```

```python
def logistic_visualizer(r=1.0, K=10, y0=1):

    def logistic(t, y):
        return r * y * (1 - y/K)

    t = np.linspace(0, 20, 500)

    sol = solve_ivp(
        logistic,
        [0,20],
        [y0],
        t_eval=t
    )

    plt.figure(figsize=(10,5))
    plt.plot(sol.t, sol.y[0], linewidth=3)

    plt.title("Logistic Growth ODE")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.grid(True)
    plt.show()
```

```python
widgets.interact(
    logistic_visualizer,
    r=(0.1,5.0,0.1),
    K=(1,100,1),
    y0=(0.1,50,0.1)
);
```
